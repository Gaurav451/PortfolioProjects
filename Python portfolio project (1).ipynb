```python
# Import Libraries

import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

pd.options.mode.chained_assignment = None

df = pd.read_csv(r'C:\Users\gaurav\Downloads\movies.csv')
```

```python
# A quick glimpse at first 5 rows of data

df.head()
```

```python
# Start cleaning data

# checking for missing values using mean method

for col in df.columns :
    pct_null = np.mean(df[col].isnull())
    print('{} - {}'.format(col,pct_null))

```

```python
# checking the data types and changing data types accordingly

df['budget'] = df['budget'].astype('Int64')

df['gross'] = df['gross'].astype('Int64')

df['runtime'] = df['runtime'].astype('Int64')
```

```python
df
```

```python
# Extracting year from released column

df['YearExtract'] = df['released'].str.extract(pat='([0-9]{4})').astype('Int64')
```

```python
df
```

```python
df = df.sort_values(by = ['gross'], inplace = False, ascending = False)
```

```python
# Display entire data

pd.set_option('display.max_rows', None)
```

```python
df.head()
```

```python
# dropping duplicates

# by a specific column

df['writer'] = df['writer'].drop_duplicates().sort_values(ascending = False)

# dropping duplicates in entire column

df = df.drop_duplicates()
```

```python
# checking for correlation

# scatter plot of budget vs gross(as budget increases, the production expects the gross(revenue) to go up)

plt.scatter(x = df['budget'], y = df['gross'])

plt.title('Budget vs Gross Earnings')

plt.xlabel('Budget')

plt.ylabel('Gross')

plt.show()
```

```python
# For determining correlation, we are going to plot a 'regplot', or 'Regression Plot'

sns.regplot(x='budget', y='gross', data=df, scatter_kws={"color" : "Green"}, line_kws={"color" : "Blue"})
```

```python
# determining correlation

# using pearson method

df.corr()

# using kenall method

df.corr(method="kendall")

#using spearman method

df.corr(method="spearman")
```

```python
# Higher correlation between budget and gross

# Plotting the heatmap of above correlation matrix

corr_matrix = df.corr()

sns.heatmap(corr_matrix, annot=True)

plt.title('Correlation Matric for numeric features')

plt.xlabel('Movie Features')

plt.ylabel('Movie Features')

plt.show()
```
